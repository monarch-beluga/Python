```python
import geopandas as gpd

gdf = gpd.read_file("坑塘_水库.shp")

shp = geemap.gdf_to_ee(gdf)
Map.addLayer(shp, {}, "shp")
```

```python
Map.add_basemap("HYBRID")
```

```python
featrues = Map.draw_features
```

```python
f = ee.FeatureCollection(featrues)
```

```python
gdf1 = geemap.ee_to_gdf(f)
```

```python
gdf1
```

```python
def polygon_to_coords_str(polygon):
    # 获取外环坐标
    return ";".join([f"{x:.6f},{y:.6f},40" for x, y in polygon.exterior.coords])


gdf1["feature"] = gdf1['geometry'].apply(polygon_to_coords_str)
```
