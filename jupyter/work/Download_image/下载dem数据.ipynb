```python
images = ee.ImageCollection("COPERNICUS/S2_HARMONIZED") \
    .filterBounds(roi)\
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))\
    .filterDate('2025', '2026')
```

```python
image = images.median().clip(roi)
rgbVis = {
  'min': 0,
  'max': 3000,
  'bands': ['B4', 'B3', 'B2'],
}

Map.addLayer(image, rgbVis, 'RGB');
```

```python
geemap.download_ee_image(image.select(['B2', 'B3', 'B4', 'B5']), r'image.tif', scale=10, region=roi, crs="epsg:3857")
```
